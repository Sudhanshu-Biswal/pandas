```python
#import titanic
df = pd.read_csv("/Users/sudhanshubiswal/Downloads/tested.csv")
df
```

```python
#common functions
```

```python
#lower/upper/capitalize/title
df['Name'].str.lower()
df['Name'].str.capitalize()
df['Name'].str.title()


#len
df["Name"][df["Name"].str.len() ==63].values[0]
#strip
df["Name"].str.strip()
```

```python
#split -> get
df["last_name"] = df["Name"].str.split(",").str.get(0)
df.head(2)
```

```python
df[["title","firstname"]]=df["Name"].str.split(",").str.get(1).str.strip().str.split(" ",n=1,expand=True)
df.head()
```

```python
df["title"].value_counts()
```

```python
# replace
df["title"] = df["title"].str.replace("Ms","Miss")
```

```python
df["title"].value_counts()
```

```python
#filtering
#Startswith/endswith
df[df["firstname"].str.startswith('A')]
df[df["firstname"].str.endswith('A')]

#isdigit/isalpha
df[df["firstname"].str.isdigit()]
```

```python
# applying regex
#contains
#search join -> both case
df[df["firstname"].str.contains("john",case=False)]
```

```python
#find lastname with start and char vowel
df[df["last_name"].str.contains("^[aeiouAEIOU].+[aeiouAEIOU]$")]
```
